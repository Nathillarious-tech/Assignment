```python
#standardize text
df["Product_Name"] = df["Product_Name"].str.strip().str.title()
df
```

```python
df["Category"] = df["Category"].str.strip().str.title()
df
```

```python
df["Supplier"] = df["Supplier"].str.strip().str.title()
df
```

```python
#fix data types
def Quantity(value):
    if value == 'twenty-seven':
        return 27
    if value == 'fourty-five':
        return 45
    elif isinstance(value, (int, float)):
        return value
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else:
       return np.nan
df['Quantity'] = df['Quantity'].apply(Quantity)
df['Quantity']
```

```python
df
```

```python
def Reorder_Level(value):
    if value == 'twenty-one':
        return 21
    if value == 'Nineteen':
        return 19
    elif isinstance(value, (int, float)):
        return value
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else:
       return np.nan
df['Reorder_Level'] = df['Reorder_Level'].apply(Reorder_Level)
df['Reorder_Level']
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Quantity"]
```

```python
df["Unit_Price"] = pd.to_numeric(df["Unit_Price"], errors="coerce")
df["Unit_Price"]
```

```python
df["Last_Update"] = pd.to_datetime(df["Last_Update"], errors="coerce")
df["Last_Update"]
```

```python
df
```

```python
#handle missing values
df["Product_Name"].fillna("Unknown", inplace=True)
df["Product_Name"]
```

```python
df["Supplier"].fillna("Unknown", inplace=True)
df["Supplier"]
```

```python
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
df["Quantity"]
```

```python
df["Unit_Price"].fillna(df["Unit_Price"].median(), inplace=True)
df
```

```python
#remove unrealistic values 
df = df[(df["Quantity"] >= 0) & (df["Unit_Price"] > 0) & (df["Unit_Price"] < 10000)]
df
```

```python
#standardize dates
df["Last_Update"] = pd.to_datetime(df["Last_Update"], errors="coerce")
df
```

```python
#products needing restocking
restock = df[df["Quantity"] < df["Reorder_Level"]][["Product_ID","Product_Name","Quantity", "Unit_Price", "Reorder_Level"]]
restock
```

```python
# Average unit price per category 
avg_price = df.groupby("Category")["Unit_Price"].mean() 
avg_price
```

```python
# Supplier with most products
supplier_count = df["Supplier"].value_counts().idxmax()
"Top supplier:", supplier_count
```

```python
# Top 5 most expensive products
top_expensive = (df.sort_values("Unit_Price", ascending=False)).head(5)
top_expensive
```
